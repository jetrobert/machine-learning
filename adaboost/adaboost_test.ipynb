```python
import numpy as np
import adaboost
```

# load data

```python
datMat, classLabels = adaboost.loadSimpData()
```

# decision stump

```python
D = np.mat(np.ones((5,1))/5)
adaboost.buildStump(datMat, classLabels, D)
```

```python
classifierArray = adaboost.adaBoostTrainDS(datMat, classLabels, 9)
print(classifierArray)
```

# adaboost classifier

```python
datArr, labelArr = adaboost.loadSimpData()
classifierArr = adaboost.adaBoostTrainDS(datArr, labelArr, 30)
adaboost.adaClassify([0, 0], classifierArr[0])
```

```python
adaboost.adaClassify([[5, 5], [0, 0]], classifierArr[0])
```

# load horse colic data

```python
datArr, labelArr = adaboost.loadDataSet('horseColicTraining2.txt')
classifierArray = adaboost.adaBoostTrainDS(datArr, labelArr, 10)
prediction = adaboost.adaClassify(datArr, classifierArray[0])
```

```python
errArr = np.mat(np.ones((299,1)))
errNums = errArr[prediction != np.mat(labelArr).T].sum()
print(errNums)
print('error rate: ', errNums/errArr.shape[0])
```

# test

```python
testArr, testLabelArr = adaboost.loadDataSet('horseColicTest2.txt')
prediction10 = adaboost.adaClassify(testArr, classifierArray[0])
```

```python
errArr = np.mat(np.ones((67,1)))
errNums = errArr[prediction10 != np.mat(testLabelArr).T].sum()
print(errNums)
print('error rate: ', errNums/errArr.shape[0])
```

# plot roc curve

```python
datArr, labelArr = adaboost.loadDataSet('horseColicTraining2.txt')
classifierArray, aggClassEst = adaboost.adaBoostTrainDS(datArr, labelArr, 10)
adaboost.plotROC(aggClassEst.T, labelArr)
```
