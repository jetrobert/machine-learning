# replace Nan with mean

```python
dataMat = pca.replaceNanWithMean()
meanVals = np.mean(dataMat, axis=0)
meanRemoved = dataMat - meanVals
covMat = np.cov(meanRemoved, rowvar=0)
eigVals, eigVects = np.linalg.eig((np.mat(covMat)))
print(eigVals)
```
